```python
import pandas as pd

url = "https://media.githubusercontent.com/media/KhalPrawira/Big-Data_Assignment/refs/heads/main/Dataset/Car%20Accident%202005-2015/Accidents0515.csv"
df = pd.read_csv(url)

display(df.head())
```

```python
print("Shape of the DataFrame:", df.shape)
print("\nData Types:\n", df.dtypes)

missing_values = df.isnull().sum()
print("\nMissing Values per column:\n", missing_values)
```

```python
df = df.drop('LSOA_of_Accident_Location', axis=1)

missing_values = df.isnull().sum()
print("\nMissing Values per column:\n", missing_values)
```

```python
df_cleaned = df.dropna()
display(df_cleaned.head())
print("\nShape of the DataFrame after dropping missing values:", df_cleaned.shape)
```

```python
missing_values = df_cleaned.isnull().sum()
print("\nMissing Values per column:\n", missing_values)

print("\nSummary Statistics for Numerical Features:\n", df_cleaned.describe())
```

```python
duplicate_rows = df_cleaned[df_cleaned.duplicated()]
print("\nJumlah baris duplikat:", duplicate_rows.shape[0])
```

# Task
Perform data cleaning and exploratory data analysis (EDA) on the provided dataset. The cleaning steps should include dropping the 'LSOA_of_Accident_Location' column, handling missing values, and checking for duplicate rows. The EDA should involve converting 'Date' and 'Time' columns to datetime, analyzing categorical and numerical variables, exploring relationships between variables, creating new features if necessary, and visualizing the data.

## Konversi tipe data

### Subtask:
Mengubah kolom 'Date' dan 'Time' ke tipe data datetime untuk memungkinkan analisis berbasis waktu.


**Reasoning**:
Combine 'Date' and 'Time' columns, convert to datetime, drop original columns, and display data types.



```python
df_cleaned['DateTime'] = df_cleaned['Date'] + ' ' + df_cleaned['Time']
df_cleaned['DateTime'] = pd.to_datetime(df_cleaned['DateTime'], format='%d/%m/%Y %H:%M')
df_cleaned = df_cleaned.drop(['Date', 'Time'], axis=1)
display(df_cleaned.dtypes)
```

## Analisis variabel kategorikal

### Subtask:
Memeriksa nilai unik dan distribusinya untuk kolom kategorikal (misalnya, `Police_Force`, `Day_of_Week`, `Road_Type`, dll.) untuk mengidentifikasi potensi masalah data atau nilai yang perlu ditangani (-1 mungkin mewakili data yang hilang atau tidak diketahui).


**Reasoning**:
Identify categorical columns and check their unique values and distribution, specifically looking for -1 which might indicate missing or unknown data.



```python
categorical_cols = [
    'Police_Force', 'Day_of_Week', 'Road_Type', 'Speed_limit', 'Junction_Detail',
    'Junction_Control', 'Pedestrian_Crossing-Human_Control', 'Pedestrian_Crossing-Physical_Facilities',
    'Light_Conditions', 'Weather_Conditions', 'Road_Surface_Conditions',
    'Special_Conditions_at_Site', 'Carriageway_Hazards', 'Urban_or_Rural_Area',
    'Did_Police_Officer_Attend_Scene_of_Accident'
]

for col in categorical_cols:
    print(f"\nColumn: {col}")
    print("Unique values:", df_cleaned[col].unique())
    print("Value counts:\n", df_cleaned[col].value_counts())
```

## Analisis variabel numerik

### Subtask:
Memeriksa distribusi (histogram, box plot) dan outlier untuk kolom numerik (misalnya, `Number_of_Vehicles`, `Number_of_Casualties`, `Speed_limit`, dll.).


**Reasoning**:
Identify the numerical columns and exclude identifier and location columns for distribution analysis.



```python
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns.tolist()
exclude_cols = ['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude', 'Latitude']
numerical_cols = [col for col in numerical_cols if col not in exclude_cols]

print("Numerical columns for distribution analysis:", numerical_cols)
```

**Reasoning**:
Generate histograms for the identified numerical columns to visualize their distributions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df_cleaned[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

**Reasoning**:
Generate box plots for the identified numerical columns to visualize their distributions and identify potential outliers.



```python
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(y=df_cleaned[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()
```

## Analisis hubungan antar variabel

### Subtask:
Menjelajahi hubungan antara variabel yang berbeda, misalnya, antara keparahan kecelakaan (`Accident_Severity`) dengan kondisi jalan, kondisi cuaca, atau waktu.


**Reasoning**:
Calculate and display the correlation matrix for numerical columns and filter for 'Accident_Severity' to examine its relationships with other numerical variables.



```python
numerical_cols_for_corr = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

correlation_matrix = df_cleaned[numerical_cols_for_corr].corr()
display(correlation_matrix)

accident_severity_corr = correlation_matrix['Accident_Severity'].sort_values(ascending=False)
print("\nCorrelation with Accident_Severity:\n", accident_severity_corr)
```

**Reasoning**:
Create visualizations to explore the relationship between 'Accident_Severity' and key categorical and numerical variables like 'Road_Surface_Conditions', 'Weather_Conditions', 'Light_Conditions', and 'Speed_limit' identified in the correlation analysis.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize Accident Severity vs Road Surface Conditions (Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Road_Surface_Conditions', y='Accident_Severity', data=df_cleaned)
plt.title('Accident Severity vs Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Accident Severity')
plt.show()

# Visualize Accident Severity vs Weather Conditions (Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Weather_Conditions', y='Accident_Severity', data=df_cleaned)
plt.title('Accident Severity vs Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Accident Severity')
plt.show()

# Visualize Accident Severity vs Light Conditions (Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Light_Conditions', y='Accident_Severity', data=df_cleaned)
plt.title('Accident Severity vs Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Accident Severity')
plt.show()

# Visualize Accident Severity vs Speed Limit (Numerical/Categorical)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Speed_limit', y='Accident_Severity', data=df_cleaned)
plt.title('Accident Severity vs Speed Limit')
plt.xlabel('Speed Limit')
plt.ylabel('Accident Severity')
plt.show()
```

**Reasoning**:
Analyze the relationship between 'Accident_Severity' and time-related features by extracting hour, day of week, month, and year from 'DateTime' and then visualizing these relationships using appropriate plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

df_cleaned['Accident_Hour'] = df_cleaned['DateTime'].dt.hour
df_cleaned['Accident_Month'] = df_cleaned['DateTime'].dt.month
df_cleaned['Accident_Year'] = df_cleaned['DateTime'].dt.year

# Visualize Accident Severity vs Hour of Day
plt.figure(figsize=(12, 6))
sns.barplot(x='Accident_Hour', y='Accident_Severity', data=df_cleaned)
plt.title('Average Accident Severity by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Accident Severity')
plt.show()

# Visualize Accident Severity vs Day of Week
plt.figure(figsize=(10, 6))
sns.barplot(x='Day_of_Week', y='Accident_Severity', data=df_cleaned)
plt.title('Average Accident Severity by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Accident Severity')
plt.show()

# Visualize Accident Severity vs Month
plt.figure(figsize=(10, 6))
sns.barplot(x='Accident_Month', y='Accident_Severity', data=df_cleaned)
plt.title('Average Accident Severity by Month')
plt.xlabel('Month')
plt.ylabel('Average Accident Severity')
plt.show()

# Visualize Accident Severity vs Year
plt.figure(figsize=(10, 6))
sns.barplot(x='Accident_Year', y='Accident_Severity', data=df_cleaned)
plt.title('Average Accident Severity by Year')
plt.xlabel('Year')
plt.ylabel('Average Accident Severity')
plt.show()
```

### Subtask:
Berdasarkan analisis, mungkin berguna untuk membuat fitur baru, seperti bulan atau tahun dari kolom 'Date', atau menggabungkan kolom tertentu.

## Visualisasi data

### Subtask:
Membuat visualisasi yang relevan untuk mengkomunikasikan temuan dari langkah-langkah di atas.


**Reasoning**:
Create a correlation heatmap of the numerical columns to visualize the relationships between them.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_for_corr = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df_cleaned[numerical_cols_for_corr].corr()

plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

**Reasoning**:
Visualize key relationships between variables using appropriate plots to illustrate the findings from the correlation analysis and previous steps.



```python
# Visualize Accident Severity vs Number of Vehicles (Numerical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number_of_Vehicles', y='Accident_Severity', data=df_cleaned)
plt.title('Accident Severity vs Number of Vehicles')
plt.xlabel('Number of Vehicles')
plt.ylabel('Accident Severity')
plt.show()

# Visualize Number of Casualties vs Number of Vehicles (Numerical)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Number_of_Vehicles', y='Number_of_Casualties', data=df_cleaned)
plt.title('Number of Casualties vs Number of Vehicles')
plt.xlabel('Number of Vehicles')
plt.ylabel('Number of Casualties')
plt.show()
```

**Reasoning**:
Visualize the distribution of key categorical variables using count plots to show the frequency of each category.



```python
categorical_cols_for_count = [
    'Day_of_Week', 'Road_Type', 'Weather_Conditions', 'Light_Conditions', 'Speed_limit',
    'Urban_or_Rural_Area', 'Accident_Severity'
]

for col in categorical_cols_for_count:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df_cleaned)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
